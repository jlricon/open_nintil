```python
import pandas as pd

import seaborn as sns

sns.set_context("talk")
import matplotlib.pyplot as plt
import statsmodels

import statsmodels.formula.api as smf
import numpy as np

plt.style.use("ggplot")
```

```python
def makefig(base, ols, gbm,gbm2):
    pd.Series(
        {"Counterfactual change (OLS)": ols, 
         "Counterfactual change (GBM, SLC)": gbm,
        "Counterfactual change (GBM, Utah)":gbm2}
    ).plot.barh(figsize=(12, 6))
    plt.title(f"Baseline:{base}", size=30)
    plt.xlabel("Change from baseline", size=30)
    plt.tight_layout()
    plt.savefig("figures/alcohol.png")


makefig(5.47, 0.022278, 0,0.644192)
```
